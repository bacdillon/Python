# Visualising ND data

Human brains are not the best at visualising higher dimensions. To understand high dimensional data, using libraries like `sklearn` have many algorithms for finding structure in high dimensional data and showing those relationships in lower dimensional plots.

Heart data FROM (https://www.kaggle.com/ronitf/heart-disease-uci)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
from sklearn import manifold

df = pd.read_csv("heart.csv")
target = df["target"].to_numpy()
df = df[["age", "sex", "cp", "trestbps", "chol", "thalach"]]
df.head()
```

## Scatter Matrix

This is the easiest and most commonly used way of visualising ND data. It works by breaking it down to many 2D surfaces.

```python
pd.plotting.scatter_matrix(df);
```

## Correlation Matrix

A very quick view of how much each parameter (column) is correlated with another. 

```python
df.corr()
```

```python
ax = sb.heatmap(df.corr(), annot=True, fmt="0.2f", square=True);
```
